```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# cargar dataset
df_1 = pd.read_csv('C:\\TFG\RawData\\Metrics\\bert_version_40_epochs.csv')
```

```python
df_1.plot(x='Step', y='Value', label='BETO', color='blue', title='BETO', xlabel='Epoch', ylabel='Valor')
plt.savefig('C:\\TFG\\Metrics\\beto_epochs_value.png')
plt.show()
```

```python
df_1 = pd.read_csv('C:\\TFG\\RawData\\Metrics\\bert_version_40_test_accuracy.csv')
df_2 = pd.read_csv('C:\\TFG\\RawData\\Metrics\\bert_version_40_val_accuracy.csv')
df_2
```

```python
plt.scatter(df_1['Step'], df_1['Value'], color='blue')
plt.scatter(df_2['Step'], df_2['Value'], color='red')
plt.title('BETO Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy') 
plt.legend(['Test', 'Validation']) 
plt.savefig('C:\\TFG\\Metrics\\beto_accuracy.png')
```

```python
df_1 = pd.read_csv('C:\\TFG\\RawData\\Metrics\\bert_version_40_test_loss.csv')
df_2 = pd.read_csv('C:\\TFG\\RawData\\Metrics\\bert_version_40_val_loss.csv')
```

```python
# plto df_1 scatter plot
plt.scatter(df_1['Step'], df_1['Value'], color='blue')
plt.scatter(df_2['Step'], df_2['Value'], color='red')
plt.title('BETO Test Loss')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')

# Aadir etiqueta de valor
for i in range(len(df_1['Value'])):
    plt.text(df_1['Step'][i], df_1['Value'][i], df_1['Value'][i])     
plt.legend(['Test', 'Validation'])
# Aadir etiqueta de valor
for i in range(len(df_2['Value'])):
    plt.text(df_2['Step'][i], df_2['Value'][i], df_2['Value'][i])   
plt.savefig('C:\\TFG\\Metrics\\beto_loss.png')
plt.show()
```
