```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# cargar dataset
df_1 = pd.read_csv('C:\\TFG\\RawData\\Metrics\\berta-ca_version_8_epochs.csv')
```

```python
df_1.plot(x='Step', y='Value', label='RoBERTa-CA', color='blue', title='RoBERTa_ca', xlabel='Epoch', ylabel='Valor')
# a√±adir etiquetas de valores
plt.savefig('C:\\TFG\\Metrics\\roberta_epochs_value.png')
plt.show()
```

```python
df_1 = pd.read_csv('C:\\TFG\\RawData\\Metrics\\berta-ca_version_8_test_accuracy.csv')
df_2 = pd.read_csv('C:\\TFG\\RawData\\Metrics\\berta-ca_version_8_val_accuracy.csv')
df_2
```

```python
# plto df_1 scatter plot
plt.scatter(df_1['Step'], df_1['Value'], color='blue')
plt.scatter(df_2['Step'], df_2['Value'], color='red')
plt.title('RoBERTa Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
# A√±adir etiqueta de valor
for i in range(len(df_1['Value'])):
    plt.text(df_1['Step'][i], df_1['Value'][i], df_1['Value'][i])     
plt.legend(['Test', 'Validation'])
# A√±adir etiqueta de valor
for i in range(len(df_2['Value'])):
    plt.text(df_2['Step'][i], df_2['Value'][i], df_2['Value'][i])   
plt.savefig('C:\\TFG\\Metrics\\roberta_accuracy.png')
```
